```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def draw(g):
  pos = nx.spring_layout(g)
  nx.draw(g, pos, \
          arrows=True, \
          with_labels=True, \
          nodelist=g.nodes(), \
          style='solid', \
          edge_color='k', \
          width=1.5, \
          node_color='y', \
          alpha=0.85)
  plt.show()
```

```python
# Datos ingresados en formato matriz de arcos y pesos
g_data = [(1,2,3),
          (1,3,5),
          (2,5,10),
          (2,4,1),
          (3,4,3),
          (3,7,4),
          (4,6,2),
          (4,7,7),
          (5,8,2),
          (5,4,3),
          (6,5,1),
          (6,8,5),
          (7,6,3),
          (7,8,5)
          ]

G = nx.DiGraph()
G.add_weighted_edges_from(g_data)
```

```python
draw(G)
```

```python
rc = nx.single_source_dijkstra(G, 1, 8, weight="weight")
rc
```
